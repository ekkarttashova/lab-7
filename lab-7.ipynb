```python
import pandas as pd
import numpy as np
import math 
from matplotlib.animation import PillowWriter
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("data2.csv", index_col = 1)
```

```python
plt.hist(df['ph'], color='skyblue', edgecolor='black',
         bins = 30)
plt.xlabel('Значение ph')
plt.ylabel('Частота')
plt.title('Гистограмма')
```

```python
x = np.linspace(-2*math.pi, 2*math.pi, 100)
y = np.linspace(-2*math.pi, 2*math.pi, 100)
x, y = np.meshgrid(x, y)


z = np.sin(x) * np.cos(y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

def update(frame):
    y = np.sin(x + frame * 0.1)
    line.set_ydata(y)
    return line,

animation = FuncAnimation(fig, update, frames=100)
writer = PillowWriter(fps=10)
animation.save('sin_animation.gif', writer=writer)
plt.title('Анимированный график y = sin(x)')
plt.show()
```
